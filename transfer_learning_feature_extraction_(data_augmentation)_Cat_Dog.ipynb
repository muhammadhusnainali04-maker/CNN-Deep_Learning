```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout,BatchNormalization
from keras.applications.vgg16 import VGG16
from keras.applications.vgg19 import VGG19
```

```python
import zipfile

zip_ref = zipfile.ZipFile('/content/CAT-DOG/training_set.zip', 'r')
zip_ref.extractall('/content/CAT-DOG')
zip_ref.close()

zip_ref = zipfile.ZipFile('/content/CAT-DOG/test_set.zip', 'r')
zip_ref.extractall('/content/CAT-DOG')
zip_ref.close()
```

```python
conv_base =VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(150,150,3)
)
```

```python
conv_base.summary()
#
```

```python
model=Sequential()

model.add(conv_base)
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
conv_base.trainable=False
```

```python
from keras.preprocessing.image import ImageDataGenerator , array_to_img, img_to_array, load_img
```

```python
batchSize=32
train_datagen=ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen=ImageDataGenerator(rescale=1./255)

train_ganerator=train_datagen.flow_from_directory(
     '/content/CAT-DOG/training_set',
      target_size=(150,150),
      batch_size=batchSize,
      class_mode='binary'
)

validation_generator=test_datagen.flow_from_directory(
     '/content/CAT-DOG/testing_set',
      target_size=(150,150),
      batch_size=batchSize,
      class_mode='binary'
)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
    )

history = model.fit_generator(
    train_ganerator,
    epochs=10,
    validate_data=validation_generator
    )
```
